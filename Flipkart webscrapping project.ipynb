```python
from bs4 import BeautifulSoup as bs
import requests as r
import pandas as pd
```

```python
url="https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
webpage= r.get(url)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
soup= bs(webpage.content,"html.parser")
```

```python
soup
```

```python
links= soup.find('div','DOjaWF gdgoEp').find_all('div','tUxRFH')
```

```python
len(links)
```

```python
links[5].find('div','KzDlHZ')
```

```python
Title=[]
Orignal_Price=[]
Dis_Price=[]
Rating=[]
for x in range(len(links)):
    Title.append(links[x].find('div','KzDlHZ').text)
    Orignal_Price.append(links[x].find('div','yRaY8j ZYYwLA').text)
    Dis_Price.append(links[x].find('div','Nx9bqj _4b5DiR').text)
    Rating.append(links[x].find('div','XQDdHH').text)             
```

```python
df=pd.DataFrame({'Title':Title,'Orignal_Price':Orignal_Price,'Dis_Price':Dis_Price,'Rating':Rating})
```
